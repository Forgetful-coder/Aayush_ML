```python
import pandas as pd
```

```python
from sklearn import datasets
```

```python
iris= datasets.load_iris()
```

```python
iris
```

```python
X = iris.data
y = iris.target
X,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state=42)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
tree_regressor = DecisionTreeClassifier().fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,8))
tree.plot_tree(tree_regressor,filled=True)
```

After height 2 there is no requirement for splitting hence we will do post pruning

```python
tree_regressor = DecisionTreeClassifier(max_depth=2).fit(X_train,y_train)
```

```python
plt.figure(figsize=(15,8))
tree.plot_tree(tree_regressor,filled=True)
```

```python
y_pred = tree_regressor.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

###PrePruning done by hyperparameter optimization

```python
#DecisionTree_PrePruning
params = {'criterion':['gini','entropy','log_loss'],
          'splitter':['best','random'],
          'max_depth':[1,2,4,5,7],
          'max_features':['auto','sqrt','log2'],
          'ccp_alpha':[1.0,2.0,0.0,3.0]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tree_classifier=DecisionTreeClassifier()
cv = GridSearchCV(tree_classifier,param_grid=params,cv=5,scoring='accuracy')
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
